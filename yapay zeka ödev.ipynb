```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub

# 1. Veriyi indir (Bu komut veriyi Kaggle sunucusundan çeker)
path = kagglehub.dataset_download("momanyc/museum-collection")

# 2. İndirilen klasörün içindeki CSV dosyasını oku
# Genellikle 'artworks.csv' isminde olur
df = pd.read_csv(f"{path}/artworks.csv")

# 3. İlk 5 satırı görerek kontrol et
df.head()
```

```python
df.isnull().sum()
```

```python
# Bölümlere (Department) göre eser sayısını görselleştirelim
plt.figure(figsize=(10,6))
df['Department'].value_counts().plot(kind='barh', color='skyblue')
plt.title('Bölümlere Göre Eser Dağılımı')
plt.xlabel('Eser Sayısı')
plt.show()
```

```python
df.info()
```

```python
# Müzedeki eser türlerinin (Classification) dağılımını görelim
plt.figure(figsize=(10,6))
df['Classification'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('En Yaygın 10 Eser Türü')
plt.ylabel('Eser Sayısı')
plt.xticks(rotation=45)
plt.show()
```
